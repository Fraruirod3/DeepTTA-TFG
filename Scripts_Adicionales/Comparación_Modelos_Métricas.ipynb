```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame con las métricas globales de los cuatro modelos tomadas a partir de los resultados
data = {
    'Modelo': [
        'A (head 10000)',
        'B (head 40000)',
        'C (ByCancerEquilibrado)',
        'D (ByDrugEquilibrado)'
    ],
    'MSE':     [1.2280, 1.1232, 1.1459, 1.1941],
    'RMSE':    [1.1081, 1.0598, 1.0705, 1.0928],
    'Pearson': [0.9414, 0.9404, 0.9261, 0.9239],
    'Spearman':[0.9311, 0.9232, 0.8941, 0.8918],
    'CIndex':  [0.8896, 0.8840, 0.8632, 0.8624]
}
df_metrics = pd.DataFrame(data)

# Lista de métricas a visualizar con sus límites para "zoom"
limits = {
    'MSE':      (1.10, 1.25),
    'RMSE':     (1.05, 1.25),
    'Pearson':  (0.92, 0.95),
    'Spearman': (0.88, 0.94),
    'CIndex':   (0.85, 0.90)
}

# Para cada métrica, se genera una gráfica de barras comparativa más "zoomed-in"
for metric, (ymin, ymax) in limits.items():
    plt.figure(figsize=(6, 4))
    bars = plt.bar(df_metrics['Modelo'], df_metrics[metric],
                   color=['skyblue', 'lightgreen', 'salmon', 'plum'])
    plt.ylim(ymin, ymax)
    plt.ylabel(metric)
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Comparativa de {metric} entre Modelos')
    plt.tight_layout()
    plt.show()

```
